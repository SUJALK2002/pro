```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

# Activation Functions
# Linear Function
linear = x

# Binary Step Function
binary = np.where(x >= 0, 1, 0)

# Sigmoid Function
sigmoid = 1 / (1 + np.exp(-x))

# ReLU Function
relu = np.maximum(0, x)

# Tanh Function
tanh = np.tanh(x)

# Softmax Function
softmax = np.exp(x) / np.sum(np.exp(x))

# Plotting
plt.figure(figsize=(10,8))

plt.subplot(3,2,1)
plt.plot(x, linear)
plt.title("Linear")

plt.subplot(3,2,2)
plt.plot(x, binary)
plt.title("Binary")

plt.subplot(3,2,3)
plt.plot(x, sigmoid)
plt.title("Sigmoid")

plt.subplot(3,2,4)
plt.plot(x, relu)
plt.title("ReLU")

plt.subplot(3,2,5)
plt.plot(x, tanh)
plt.title("Tanh")

plt.subplot(3,2,6)
plt.plot(x, softmax)
plt.title("Softmax")

plt.tight_layout()
plt.show()
```
