```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Take 2 features
X = iris.data[:, [0, 2]]

# Output labels
y = iris.target

# Convert into binary classification
y = np.where(y == 0, 0, 1)

# Initialize weights and bias
w = np.zeros(2)
b = 0

# Training settings
lr = 0.1
epochs = 30

# Perceptron prediction function
def predict(x):

    return np.where(np.dot(x, w) + b >= 0, 1, 0)

# Training
for _ in range(epochs):

    for i in range(len(X)):

        y_pred = predict(X[i])

        error = y[i] - y_pred

        w = w + lr * error * X[i]

        b = b + lr * error

# Create mesh grid
x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5

y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.05),
    np.arange(y_min, y_max, 0.05)
)

# Predict all grid points
Z = predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)

# Plot decision region
plt.contourf(xx, yy, Z, alpha=0.4)

# Plot data points
plt.scatter(X[:,0], X[:,1], c=y)

plt.xlabel("Sepal Length")

plt.ylabel("Petal Length")

plt.title("Perceptron Decision Regions")

plt.show()

# Input:
# Input consists of Iris flower dataset containing sepal length and petal length features used for binary classification using perceptron algorithm.
# Output:
# Program trains perceptron model on Iris dataset and displays graphical decision regions separating two flower classes.
# import numpy as np
# Imports numpy library used for arrays, mathematical operations and matrix calculations.
# import matplotlib.pyplot as plt
# Imports matplotlib library used for plotting graphs and visualizing decision regions.
# from sklearn.datasets import load_iris
# Imports Iris dataset from sklearn datasets module.
# iris = load_iris()
# Loads Iris flower dataset into iris variable.
# X = iris.data[:, [0, 2]]
# Selects only two features from dataset. Feature 0 is sepal length and feature 2 is petal length.
# y = iris.target
# Stores target class labels of Iris dataset.
# y = np.where(y == 0, 0, 1)
# Converts multiclass dataset into binary classification where class 0 remains 0 and all other classes become 1.
# w = np.zeros(2)
# Initializes weight vector with zeros for two input features.
# b = 0
# Initializes bias value with zero.
# lr = 0.1
# Defines learning rate used for weight and bias updates during training.
# epochs = 30
# Defines number of training iterations over complete dataset.
# def predict(x):
# Defines perceptron prediction function.
# return np.where(np.dot(x, w) + b >= 0, 1, 0)
# Calculates weighted sum using dot product and bias. Returns 1 if result is greater than or equal to zero otherwise returns 0.
# for _ in range(epochs):
# Starts training loop for multiple epochs.
# for i in range(len(X)):
# Iterates through all training samples one by one.
# y_pred = predict(X[i])
# Predicts output for current input sample.
# error = y[i] - y_pred
# Calculates prediction error by subtracting predicted output from actual target output.
# w = w + lr * error * X[i]
# Updates weights using perceptron learning rule.
# b = b + lr * error
# Updates bias value using prediction error and learning rate.
# x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
# Calculates minimum and maximum values for x-axis with extra margin for graph plotting.
# y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
# Calculates minimum and maximum values for y-axis with extra margin.
# xx, yy = np.meshgrid(
# Creates mesh grid for plotting decision regions.
# np.arange(x_min, x_max, 0.05),
# Generates x-axis grid points with spacing 0.05.
# np.arange(y_min, y_max, 0.05)
# Generates y-axis grid points with spacing 0.05.
# )
# Ends mesh grid creation.
# Z = predict(np.c_[xx.ravel(), yy.ravel()])
# Predicts output class for every point in mesh grid.
# Z = Z.reshape(xx.shape)
# Reshapes predicted outputs into grid shape for plotting.
# plt.contourf(xx, yy, Z, alpha=0.4)
# Plots colored decision regions using predicted grid outputs.
# plt.scatter(X[:,0], X[:,1], c=y)
# Plots actual dataset points on graph with colors according to class labels.
# plt.xlabel("Sepal Length")
# Sets x-axis label as Sepal Length.
# plt.ylabel("Petal Length")
# Sets y-axis label as Petal Length.
# plt.title("Perceptron Decision Regions")
# Sets graph title.
# plt.show()
# Displays final graph containing decision regions and dataset points.
# Logic Of Program:
# Step 1: Load Iris flower dataset.
# Step 2: Select two features for two-dimensional visualization.
# Step 3: Convert multiclass labels into binary classification labels.
# Step 4: Initialize perceptron weights and bias.
# Step 5: Define prediction function using weighted sum and threshold condition.
# Step 6: Train perceptron model using learning rule.
# Step 7: Update weights and bias repeatedly using prediction error.
# Step 8: Create mesh grid covering complete feature space.
# Step 9: Predict output class for all grid points.
# Step 10: Plot decision regions and dataset points on graph.
# Working Of Perceptron Decision Region:
# Perceptron learns separating boundary between two classes.
# Weighted sum and bias decide output class.
# Decision region shows areas predicted for each class.
# Different colors represent different predicted classes.
# Scatter plot displays actual dataset points.
# Viva Questions And Answers:
# Q1. What is perceptron in neural networks?
# Ans: Perceptron is simple single-layer neural network used for binary classification problems using weighted inputs and activation function.
# Q2. What is Iris dataset?
# Ans: Iris dataset contains measurements of Iris flowers and is commonly used for machine learning classification experiments.
# Q3. Why are only two features selected?
# Ans: Two features are selected for easy two-dimensional graph visualization and understanding of perceptron decision regions.
# Q4. What is binary classification?
# Ans: Binary classification is machine learning task where outputs belong to only two classes such as 0 and 1.
# Q5. What are decision regions?
# Ans: Decision regions are graph areas showing which class the perceptron predicts for different input values.
# Q6. What is decision boundary?
# Ans: Decision boundary is line separating different predicted classes generated by perceptron model during classification.
# Q7. What is activation function in perceptron?
# Ans: Activation function generates final output based on weighted sum and threshold comparison inside perceptron model.
# Q8. What is weighted sum?
# Ans: Weighted sum is total obtained after multiplying inputs with corresponding weights and adding all results together.
# Q9. What is bias in perceptron?
# Ans: Bias is additional parameter helping shift decision boundary and improving classification capability of perceptron model.
# Q10. What is learning rate?
# Ans: Learning rate controls amount of weight and bias updates during perceptron training process.
# Q11. What is prediction error?
# Ans: Prediction error is difference between actual target output and predicted output generated by perceptron.
# Q12. What is mesh grid in matplotlib?
# Ans: Mesh grid creates coordinate points over graph area used for plotting decision regions and contour plots.
# Q13. What is np.where() used for?
# Ans: np.where() checks conditions and returns values depending on whether condition becomes true or false.
# Q14. What is contourf() function?
# Ans: contourf() creates filled contour plots used for visualizing decision regions in machine learning classification graphs.
# Q15. What is scatter plot?
# Ans: Scatter plot displays individual dataset points on graph according to feature values and class labels.
# Q16. What is supervised learning?
# Ans: Supervised learning is learning method where model trains using labeled input data and corresponding outputs.
# Q17. What is epoch in perceptron training?
# Ans: Epoch represents one complete cycle where entire training dataset passes through perceptron during learning process.
# Q18. What is use of numpy in ANN?
# Ans: Numpy provides efficient array operations, vector calculations and mathematical functions required for neural network implementation.
# Q19. What is limitation of perceptron?
# Ans: Perceptron can solve only linearly separable problems and cannot handle complex nonlinear classification tasks like XOR.
# Q20. What is importance of graph visualization in ANN?
# Ans: Graph visualization helps understand model behavior, classification regions and separation boundaries learned during training process.
```
