```python
# Backpropagation Neural Network for XOR Function

import numpy as np

# XOR input
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

# XOR output
Y = np.array([[0],
              [1],
              [1],
              [0]])

# Random weights
w1 = np.random.rand(2,2)
w2 = np.random.rand(2,1)

# Learning rate
lr = 0.1

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative
def derivative(x):
    return x * (1 - x)

# Training
for i in range(10000):

    # Forward Propagation

    hidden_input = np.dot(X, w1)

    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, w2)

    output = sigmoid(final_input)

    # Error
    error = Y - output

    # Back Propagation

    d_output = error * derivative(output)

    hidden_error = d_output.dot(w2.T)

    d_hidden = hidden_error * derivative(hidden_output)

    # Update weights

    w2 += lr * hidden_output.T.dot(d_output)

    w1 += lr * X.T.dot(d_hidden)

# Final Output
print("Predicted Output:")

print(np.round(output))
```
